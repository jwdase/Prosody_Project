```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
lang_tensor = {}
languages = ['de', 'it', 'en', 'es']

for lang in languages:

    lang_tensor[lang] = torch.load(
        f'/om2/user/moshepol/prosody/data/low_pass/{lang}/spect/train/range_5_5-6_0/batch_00000_128.pt',
        weights_only = True,
        )[random.randint(0, 128)].numpy()
```

```python
plt.figure(figsize=(12, 8))
```

```python

for i, (lang, spec) in enumerate(lang_tensor.items()):
    plt.subplot(2, 2, i+1)  # 2 rows, 2 columns, i+1-th subplot
    plt.imshow(spec, aspect="auto", origin="lower", cmap="viridis")

    print(spec.shape)

    plt.title(f"Spectrogram {i+1}: {lang}")
    plt.colorbar(format="%+2.0f dB")
    plt.tight_layout()

plt.show()
```
