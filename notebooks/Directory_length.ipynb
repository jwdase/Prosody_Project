```python
from langaugedetection.data.com_voice_dir import open_files, person_to_group

# Open Validated .tsv, and then form dict 
# key : speaker id
# item : .tsv of all speaker files
language = 'en'
files = person_to_group(open_files(language, 10_000))
```

```python
from langaugedetection.data.audio_length import df_values
import pandas as pd

name = list(files.keys())[0]
person = files[name]

def make_single_string(name, maps):
    '''
    Takes in a list of URL's and adds a hypthin between them
    and forms it into a pandas datastructure
    that can be added to our csv file
    '''
    dictionary = {'id' : name}

    for val, urls in maps.items():
        dictionary[val] = [urls]

    return dictionary

def make_into_list(name, maps):
    '''
    Takes an output from df_values {range : urls}
    and turns it into a list that can be added to the
    dataframe
    '''
    insert = [name]

    for key, urls in maps.items():
        insert.append(urls)

    return insert

maps = df_values(person, language)
x = make_single_string(name, maps)

df = pd.DataFrame(x)
```

```python
index = 1
for name, urls in files.items():

    # Doesn't Enter first value
    if index == 1:
        index += 1
        continue

    df.loc[index] = make_into_list(name, df_values(urls, language))
    index += 1
```
